```python
1. Signal Classification Using Machine Learning
Problem Statement: You have a dataset of electrical signals, where each signal
represents either a &quot;sine wave&quot; or a &quot;square wave.&quot; The dataset contains the following
features for each signal: - Amplitude (in volts) , Frequency (in Hz) , Duration (in
seconds) , Peak-to-Peak Voltage (in volts) .
Your task is to develop a machine learning model that can classify each signal as
either &quot;Sine&quot; or &quot;Square&quot; based on these features using KNN. Encode labels (Sine →
0, Square → 1)

S.No Amplitude Frequency Duration Peak-to-Peak Label
1 1 50 1 2 Sine
2 1.5 60 0.5 3 Square
3 0.8 55 1.2 1.6 Sine
4 2 40 1 4 Square
5 1.2 65 0.8 2.4 Sine
6 1.7 45 1.5 3.4 Square
7 0.9 70 1 1.8 Sine
8 2.1 35 1.3 4.2 Square
9 1.1 60 0.9 2.2 Sine
10 1.8 48 1 3.6 Square
11 0.95 62 1.1 1.9 Sine
12 2.2 42 0.7 4.4 Square
13 1 58 1 2 Sine
14 1.9 46 1.2 3.8 Square
15 1.3 66 0.6 2.6 Sine
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score

data = {
    'Amplitude': [1, 1.5, 0.8, 2, 1.2, 1.7, 0.9, 2.1, 1.1, 1.8, 0.95, 2.2, 1, 1.9, 1.3],
    'Frequency': [50, 60, 55, 40, 65, 45, 70, 35, 60, 48, 62, 42, 58, 46, 66],
    'Duration': [1, 0.5, 1.2, 1, 0.8, 1.5, 1, 1.3, 0.9, 1, 1.1, 0.7, 1, 1.2, 0.6],
    'Peak-to-Peak': [2, 3, 1.6, 4, 2.4, 3.4, 1.8, 4.2, 2.2, 3.6, 1.9, 4.4, 2, 3.8, 2.6],
    'Label': ['Sine', 'Square', 'Sine', 'Square', 'Sine', 'Square', 'Sine', 'Square',
              'Sine', 'Square', 'Sine', 'Square', 'Sine', 'Square', 'Sine']
}

df = pd.DataFrame(data)
df['Label'] = df['Label'].map({'Sine': 0, 'Square': 1})

feature_names = ['Amplitude', 'Frequency', 'Duration', 'Peak-to-Peak']
X = df[feature_names]
y = df['Label']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

print("--- Model Performance ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Sine', 'Square']))

# We use a DataFrame here to avoid the "UserWarning" about feature names
new_signal_values = [[1.5, 50, 1.0, 3.0]]
new_signal_df = pd.DataFrame(new_signal_values, columns=feature_names)

new_signal_scaled = scaler.transform(new_signal_df)
prediction = knn.predict(new_signal_scaled)
result = "Square" if prediction[0] == 1 else "Sine"

print("-" * 30)
print(f"--- Test Prediction ---")
print(f"Input Features: {new_signal_values[0]}")
print(f"Predicted Class: {result}")
```
