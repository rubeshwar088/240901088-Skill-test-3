```python
Write a program to build a machine learning linear regression model to predict
the price of a house based on its features. You are given a dataset containing
information of Year and price.

Year Price(₹ Lakhs)
2010 25
2011 28
2012 30
2013 34
2014 37
2015 40
2016 43
2017 45
2018 47
2019 50
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]).reshape(-1, 1)
y = np.array([25, 28, 30, 34, 37, 40, 43, 45, 47, 50])

model = LinearRegression()
model.fit(X, y)

year_to_predict = np.array([[2020]])
predicted_price = model.predict(year_to_predict)

print(f"Predicted house price for {year_to_predict[0][0]}: ₹{predicted_price[0]:.2f} Lakhs")

plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression Line')
plt.xlabel('Year')
plt.ylabel('Price (₹ Lakhs)')
plt.title('House Price Prediction')
plt.legend()
plt.show()

print(f"Equation: Price = {model.coef_[0]:.2f} * Year + {model.intercept_:.2f}")
```
