```python
Write a program to classify a component is defective or not based on Voltage and
Current using KNN. Encode labels (Working → 1, Defective → 1)

Voltage(V) Current(A) Status
5 0.5 Working
5.5 0.4 Working
4.8 0.6 Working
6 0.3 Defective
6.5 0.2 Defective
6.2 0.25 Defective
4.9 0.55 Working
5.8 0.35 Defective
5.2 0.5 Working
6.3 0.22 Defective
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

data = {
    'Voltage': [5, 5.5, 4.8, 6, 6.5, 6.2, 4.9, 5.8, 5.2, 6.3],
    'Current': [0.5, 0.4, 0.6, 0.3, 0.2, 0.25, 0.55, 0.35, 0.5, 0.22],
    'Status': ['Working', 'Working', 'Working', 'Defective', 'Defective',
               'Defective', 'Working', 'Defective', 'Working', 'Defective']
}

df = pd.DataFrame(data)

df['Status'] = df['Status'].map({'Defective': 0, 'Working': 1})
X = df[['Voltage', 'Current']]
y = df['Status']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("--- Component Classification Model ---")
print(f"Model Accuracy: {accuracy * 100:.2f}%")

new_comp = pd.DataFrame([[5.1, 0.45]], columns=['Voltage', 'Current'])
new_comp_scaled = scaler.transform(new_comp)
prediction = knn.predict(new_comp_scaled)

result = "Working" if prediction[0] == 1 else "Defective"
print(f"\nTest Input: [5.1V, 0.45A] -> Result: {result}")
```
